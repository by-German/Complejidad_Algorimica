##Búsqueda en grafos

En términos generales, un grafo es un conjunto de vértices, también conocidos como nodos y aristas que se conforman a partir de un par de vértices, estos pueden estar orientados o no.
Los algoritmos más esenciales para el desarrollo del proyect en grafos en cuanto a busqueda serían los siguientes:
Algoritmo de búsqueda en anchura (BFS) y Algoritmo de búsqueda en
profundidad (DFS).
##Breadth First Search (BFS)


El algoritmo de búsqueda en anchura, conocido como BFS, permite al usuario recorrer los elementos de un grafo comenzando desde la raíz explorando todos los hijos del vertice. Posteriormente explorará cada uno de los hijos hasta hallar una coincidencia.
##Distributed File System (DFS)

El algoritmo de busqueda de profundidad, tambien llamado DFS permite al usuario recorrer todos los vértices de un grafo, y cuando se llegue al final, recorrer un camino de regreso, con el metodo de Backtracking, de esta forma repetir el mismo proceso con cada hermano del vertice visitado anteriormente.

# Backtracking
El Backtracking es una técnica utilizada en la programación para llevar a cabo una búsqueda sistemática por medio de todas las variables posibles en un espacio de búsqueda. Los algoritmos de Backtracking se dedican a la construcción de soluciones posibles de manera que se verifica si esta variable cumple con los requisitos de la solución o no, y con esta información se continúa con el proceso de busqueda. De esta forma, se construyen todas los posibles caminos de la solución , y se invoca recursivamente al algoritmo en las mismas.

Una breve explicación del comportamiento del backtracking es decir que consiste en enumerar de manera sistemática las alternativas de solución al problema y al probar cada una de las alternativas se guarda en memoria del resto de alternativas, esto servirá en el caso de que no se elija el camino correcto para llegar a la solución, se pueda volver a un punto para probar con otras alternativas.
 
Esta técnica es usualmente representada a modo de árbol debido a que facilita el entendimiento del mismo. El árbol se recorre en profundidad. Un camino a la solución está dada por un conjunto de vértices de tal forma que se elabora una secuencia que logra resolver el problema.

## Incorporación de Barreras en el juego

Las barreras en el juego son una funcionalidad que impide el movimiento de los jugadores en el tablero $G$ (Grafo) . Para implementar dicha funcionalidad tenemos que tener en cuenta que los jugadores en el tablero se logran mover a través de las conexionse de las aristas con los nodos. En ese sentido, colocar una barrera o impedir el paso de un jugador, desde su origen $s$ hacia un destino $u$, es simplemente desvincular un nodo de otro, o dicho de otra forma, ***remover la arista $(u, v)$*** que une dichos nodos. 

Sin embargo, la incorpación de las barreras en el juego supone otras dificultades, entre estas se encuentran:


* ***Desición de colocar una barrera***: 
En este punto, el jugador evalúa si es conveniente colocar una barrera o continuar con su camino. De tal modo que la elección le permita estar más cerca de la victoria. Para ello, el jugador debe evaluar la distancia que le falta para llegar al destino, en comparación con sus adversarios, de tal modo que, si el camino para llegar al destino es menor que el de sus adversarios, es conveniente que el jugador siga adelante.

*  ***NO Encerrar a los adversarios***: En este punto, por reestricción de las reglas del juego, el jugador al colocar una barrera, no debe de encerrar a sus adversarios. Para ello, se utilizará el concepto de **Arista de corte**, el cual evalúa que un jugador no se quede aislado en el grafo $G$,  lo cual representa que esta encerrado.

### Gerneración del Grafo:
El Quoridor, es un tablero de dimension $ 9 \times 9$, en donde los jugadores pueden desplzarse normalmente hacia arriba / abajo o izquierda / derecha, mientras tengan la posibilidad de hacerlo. Por ello, para representarlo se ha optado por hacerlo mediante la generacion de un grafo, ya que de este modo se podra simular los posibles movimientos que podra hacer el jugador en todo el partido a traves de las aristas de grafo.

A continuacón, se representa la funcion que nos permitira generar el grafo, con dimensiones arbitrarias $ n \times n $:

```
def grafo_tablero(n):   # Complejidad -> O(n^2)
    count = 0
    G = nx.DiGraph()
    for x in range(n):
        for y in range(n):
            count += 1
            if not (y == n - 1): # enlace  horizontal
                G.add_edge(count, count + 1)
                G.add_edge(count + 1, count)
            if not (x == n - 1): # enlace vertical
                G.add_edge(count, count + n)
                G.add_edge(count + n, count)
            G.nodes[count]["id"] = count
    return G
```

```
n = 5
G = grafo_tablero(n)
nx.draw(G, with_labels = True)
```

#### Implementacion del algoritmo BFS:

```
def BFS(G, s):
    for _, u in G.nodes(data = True):
        u["color"] = "blanco"
        u["distancia"] = None # infinito
        u["padre"] = None
    s["color"] = "gris"
    s["distancia"] = 0
    s["padre"] = None
    Q = deque()
    Q.append(s) # enqueue
    while not (Q == deque([])):
        u = Q.popleft()
        for _, v in G.edges(u["id"]):
            nodo = G.nodes[v]
            if nodo["color"] == "blanco":
                nodo["color"] = "gris"
                nodo["distancia"] = u["distancia"] + 1
                nodo["padre"] = u # u["id"]
                Q.append(nodo)
        u["color"] = "negro"
```

#### Implementacion del algoritmo DFS:

```
def DFS_visit(G, u):
    global tiempo
    tiempo += 1
    u["inicio"] = tiempo
    u["color"] = "gris"
    for _, v_id in G.out_edges(u["id"]): 
        v = G.nodes[v_id]
        if v["color"] == "blanco":
            v["padre"] = u
            DFS_visit(G, v)
    u["color"] = "negro"
    tiempo += 1
    u["fin"] = tiempo

def DFS(G, s):
    global tiempo
    for _, u in G.nodes(data = True): # _: id, u: data
        u["color"] = "blanco"
        u["padre"] = None
    tiempo = 0
    DFS_visit(G, s)
```

#### Implementacion del Algoritmo Djsktra:

```
def Dijkstra(Grafo, inicio, meta):
    shrt = {}  # Grabar el costo para llegar al nodo, shrt de shortpath
    nddesc = Grafo  # itera el grafo completo
    inf = 999999  # Solamente un numero grande, que sera mas grande que los pesos de los nodos
    path = []  # grabar el camino de regreso al nodo, el mas optimo

    for node in nddesc:
        shrt[node] = inf
    shrt[inicio] = 0

    while nddesc:
        min_dist = None
        for node in nddesc:
            if min_dist is None:
                min_dist = node
            elif shrt[node] < shrt[min_dist]:
                min_dist = node

        opciones = Grafo.edges(min_dist, data = "peso")
        for origen,hijos, peso in opciones:
            if peso + shrt[min_dist] < shrt[hijos]:
                shrt[hijos] = peso + shrt[min_dist]
        nddesc.remove_node(min_dist)
    NodoActual = meta
    path.insert(0, inicio)
    if shrt[meta] != inf:
        return str(shrt[meta])
```

#### Implementacion Print-Path:
Esta funcion nos permite conocer cuales son los nodos por los cuales se deben de recorrer , para llegar de un nodo origen $s$ a un nodo destino $v$, segun el algoritmo de ***busqueda de grafos*** utilizado.

```
def hallar_camino(G, s, v, camino):
    if v["id"] == s["id"]:
        camino.append(s["id"])
    elif v["padre"] == None:
        print("No existe camino de {} a {}".format(s["id"], v["id"]))
    else:
        hallar_camino(G, s, v["padre"], camino)
        camino.append(v["id"])
```

## Realizando pruebas de eficiencia en recorrido:



```
def test_recorrido(n, inicio, fin, alg):  
    global tiempo
    tiempo = 0
    G = grafo_tablero(n)
    camino = []
    origen = G.nodes[inicio] # posicion del jugador
    destino = G.nodes[fin]
    alg(G, origen)
    hallar_camino(G, origen, destino, camino)
    print(camino)
```

La siguiente funcion nos permitira medir los tiempos de carga de los algorimos respectivos.

```
import time
import pylab
import matplotlib.pyplot as plt
import sys
sys.setrecursionlimit(10**7)
```

```
def measure_time(sorting_alg, n, origen, destino ):
  global tiempo
  tiempo = 0
  G = grafo_tablero(n)
  camino = []
  start = time.time()
  sorting_alg(G, G.nodes[origen])
  hallar_camino(G, G.nodes[origen], G.nodes[destino], camino)
  end = time.time()
  return end-start
```

```
def measure_time_Dijks(Algoritm,n, origen, destino):
  global tiempo
  tiempo =0
  G = grafo_tablero(n)
  start = time.time()
  Algoritm(G, origen, destino)
  end = time.time()
  return end - start
```
